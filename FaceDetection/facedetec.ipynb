```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
frame = cv2.imread("cr7.jpg")
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
#haarcode from github
```

```python
faces=classifier.detectMultiScale(frame)
```

```python
n = len(faces)
```

```python
i=0
while(n>i):
    face = faces[i]
    x, y, w, h = face
    face = rgb[y:y+h, x:x+w]
    facerec = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,255,0),5)
    i+=1
plt.imshow(facerec)
#this code can detect multiple faces in single picture
```
